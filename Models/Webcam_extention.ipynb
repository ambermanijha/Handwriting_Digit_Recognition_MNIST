```python
import numpy as np
import cv2
import tensorflow as tf

font_scale = 1.5
font = cv2.FONT_HERSHEY_PLAIN
IMG_SIZE = 28

cap = cv2.VideoCapture(1)
cap.set(cv2.CAP_PROP_FPS, 170)
if not cap.isOpened():
    cap = cv2.VideoCapture(0)
if not cap.isOpened():
    raise IOError("Cannot open video")

model = tf.keras.models.load_model('digit_mnist.h5')  # Load your model here

cntr = 0
while True:
    ret, frame = cap.read()
    cntr = cntr + 1
    if ((cntr % 2) == 0):
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        resized = cv2.resize(gray, (28, 28), interpolation=cv2.INTER_AREA)
        newimg = tf.keras.utils.normalize(resized, axis=1)
        newimg = np.array(newimg).reshape(-1, IMG_SIZE, IMG_SIZE, 1)
        predictions = model.predict(newimg)
        status = np.argmax(predictions)
        print(status)
        print(type(status))

        x1, y1, w1, h1 = 0, 0, 175, 175
        cv2.rectangle(frame, (x1, x1), (x1 + w1, y1 + h1), (0, 0, 0), -1)
        cv2.putText(frame, str(status), (x1 + int(w1 / 5), y1 + int(h1 / 2)), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2)

        # Draw a box around the detected digit
        x2, y2, w2, h2 = 100, 100, 200, 200  # adjust these values to match your digit location
        cv2.rectangle(frame, (x2, y2), (x2 + w2, y2 + h2), (0, 255, 0), 2)

        # Remove mirror effect
        frame = cv2.flip(frame, 1)

        cv2.imshow("Handwritten Digits Recognition Tutorial", frame)

        if cv2.waitKey(2) & 0xFF == ord('q'):
            break

cap.release()
cv2.destroyAllWindows()
```
